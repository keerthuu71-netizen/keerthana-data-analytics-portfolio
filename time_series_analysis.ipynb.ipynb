```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
df = pd.read_csv("data.csv")
```

```python
df['time'] = pd.date_range(start='2020-01-01', periods=len(df), freq='D')
df.set_index('time', inplace=True)
```

```python
ts = df['sepal_length']
```

```python
plt.figure(figsize=(10,5))
plt.plot(ts, label='Sepal Length')
plt.title('Time Series Plot of Sepal Length')
plt.xlabel('Time')
plt.ylabel('Sepal Length')
plt.legend()
plt.show()
```

```python
decomposition = seasonal_decompose(ts, model='additive', period=12)
decomposition.plot()
plt.show()
```

```python
df['moving_avg'] = ts.rolling(window=10).mean()

plt.figure(figsize=(10,5))
plt.plot(ts, label='Original')
plt.plot(df['moving_avg'], label='Moving Average', color='red')
plt.legend()
plt.title('Moving Average Smoothing')
plt.show()
```
